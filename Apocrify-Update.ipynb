```python
# Import of nessery library

import pandas as pd
import numpy as np
import mysql.connector
from mysql.connector import (connection)
```

```python
def update_month(month,month1):
    
    import mysql.connector

    try:
        mydb = mysql.connector.connect(user='root', password='koreniz2',
                                     host='127.0.0.1',
                                     database='apocrifi_projec')
        cursor = mydb.cursor()

        print("Before updating a record ")
        sql_select_query = """select * from salary where id in (select id 
        where month =%s and salary_month=0 )"""
        q=(month)
        cursor.execute(sql_select_query,(q,))
        record = cursor.fetchall()
        print(record)

        # Update single record now
        sql_update_query = """Update salary set month = %s where id in (select id 
        where month =%s and salary_month=0)"""
        q=(month1,month)
        cursor.execute(sql_update_query,q)
        mydb.commit()
        print("Record Updated successfully ")

        print("After updating record ")
        sql= """select * from salary where id in (select id 
        where month =%s and salary_month=0 )"""
        q=(month1)
        cursor.execute(sql,(q,))
        record = cursor.fetchall()
        print(record)

    except mysql.connector.Error as error:
        print("Failed to update table record: {}".format(error))
    finally:
        if mydb.is_connected():
            mydb.close()
            print("MySQL connection is closed")
```

```python
p=[('01', '1'),('02', '2'),('03', '3'),('04', '4'), ('05', '5'), ('06', '6'), ('07', '7'), ('08', '8'),('09','9')]

for month,month1 in p:
    update_month(month,month1)
```
