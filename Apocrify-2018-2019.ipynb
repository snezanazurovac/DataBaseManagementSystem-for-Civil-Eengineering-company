```python
# we want to put data in table salary
# we make function salary_calculation to execute sql query true iteration 
# true different months,years,and employe_id numbers

def salary_calculation(employee_id,month,year):
    
    mycursor = mydb.cursor()
    
    # sql query to take data from two tables (employee and hours, and put in talbe salary )
    
    sql = """ select employee_id, month(work_date) ,year(work_date), salary_per_hour,
            sum(hours),sum(hours)*salary_per_hour 
            From hours 
            join employee 
            using(employee_id) 
            where employee_id=%s and month(work_date)= %s 
            and year(work_date)=%s """
    
    tip=(employee_id,month,year)
    
    # set variable in query
    mycursor.execute(sql,tip)
    
    # fetch result
    myresult= mycursor.fetchall()
    # since sql query returns row with null values,we have to prepare those row in data type
    # that can be imported into database
    
    if myresult==[(None, None, None, None, None, None)]:
        myresult=[(employee_id,month,year,0,0,0)]
    else:
        myresult=myresult
    
    return myresult
```

```python
# function to insert data from salary_calculation function 

def insert_into_salary(p):
    
    sql = """
            INSERT INTO salary (employee_id,month,year,price_hour,total_hours,salary_month)
            
            VALUES (%s,%s,%s,%s,%s,%s) """
    
    # we have to make tuple of valeus we get from salary_calculation

    q=tuple([str(p[0][i]) for i in range(6)])
    
    mycursor.execute(sql,q)

    mydb.commit()

    print(mycursor.rowcount, "record inserted.")
    
```

```python
# imput data of salary for all combination of employee_id,month,year


for year in ['2018','2019']:
    for month in['01','02','03','04','05','06','07','08','09','10','11','12']:
        for employee_id in [str(i) for i in range(1,9)]:
            
            insert_into_salary(salary_calculation(employee_id,month,year))
```
